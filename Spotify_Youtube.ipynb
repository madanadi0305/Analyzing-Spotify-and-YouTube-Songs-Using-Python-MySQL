```python
# These columns have only 2 nill values so we can replace it with mean value
df['Danceability']=df['Danceability'].fillna(df['Danceability'].mean())
df['Energy']=df['Energy'].fillna(df['Energy'].mean())
df['Key']=df['Key'].fillna(df['Key'].mean())
df['Loudness']=df['Loudness'].fillna(df['Loudness'].mean())
df['Speechiness']=df['Speechiness'].fillna(df['Speechiness'].mean())
df['Acousticness']=df['Acousticness'].fillna(df['Acousticness'].mean())
df['Instrumentalness']=df['Instrumentalness'].fillna(df['Instrumentalness'].mean())
df['Liveness']=df['Liveness'].fillna(df['Liveness'].mean())
df['Valence']=df['Valence'].fillna(df['Valence'].mean())
df['Tempo']=df['Tempo'].fillna(df['Tempo'].mean())
df['Duration_ms']=df['Duration_ms'].fillna(df['Duration_ms'].mean())
```

```python
df['Url_youtube']=df['Url_youtube'].fillna('Not Specified')
df['Title']=df['Title'].fillna("Not Specified")
df['Channel']=df['Channel'].fillna('Not Specified')
```

```python
remove_columns=['Url_spotify','Uri','Key','Url_youtube','Description']
df=df.drop(columns=remove_columns)
```

```python
df.isna().sum()
```

```python
df['Views']=df['Views'].fillna(df['Views'].mean())
df['Likes']=df['Likes'].fillna(df['Likes'].mean())
df['Comments']=df['Comments'].fillna(df['Comments'].mean())
df['Stream']=df['Stream'].fillna(df['Stream'].mean())
df['official_video']=df['official_video'].fillna(True)
df['Licensed']=df['Licensed'].fillna(True)
```

```python
df=df.rename(columns={'Duration_ms':'Duration_min'})
```
